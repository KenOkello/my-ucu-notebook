```python
#2.6
class Member(ABC):
    def __init__(self, name, id_no, savings_balance):
        self.name = name
        self.id_no = id_no
        self.savings_balance = savings_balance

    @abstractmethod
    def loan_eligibility(self):
        pass

    @staticmethod
    def get_exchange_rate():
        """Retrieve current USD→UGX exchange rate with error handling."""
        url = "https://api.exchangerate-api.com/v4/latest/USD"
        try:
            response = requests.get(url, timeout=10)
            response.raise_for_status()
            data = response.json()
            rate = data["rates"].get("UGX")
            if rate is None:
                raise KeyError("UGX rate not found in response.")
            return rate
        except requests.exceptions.RequestException as e:
            print(f"Network error: {e}")
        except ValueError:
            print("Error decoding JSON response.")
        except KeyError as e:
            print(f"Data error: {e}")
        return None


#Subclass: FarmerMember
class FarmerMember(Member):
    def loan_eligibility(self):
        return self.savings_balance * 6


# Subclass: TraderMember
class TraderMember(Member):
    def loan_eligibility(self):
        return self.savings_balance * 4


#Demonstration
farmer = FarmerMember("Okello Peter", "FM001", 150000)
trader = TraderMember("Namubiru Sarah", "TM002", 300000)

usd_rate = Member.get_exchange_rate()

if usd_rate:
    print(f"Current USD→UGX rate: {usd_rate}")
else:
    print("Exchange rate unavailable. Using UGX values only.")

print("--------------------------------------------------")

for member in [farmer, trader]:
    ugx_eligibility = member.loan_eligibility()
    usd_eligibility = ugx_eligibility / usd_rate if usd_rate else "N/A"
    print(f"{member.name} ({member.__class__.__name__})")
    print(f"  Loan Eligibility: UGX {ugx_eligibility:,.0f}")
    print(f"  Equivalent in USD: {usd_eligibility if usd_eligibility == 'N/A' else f'${usd_eligibility:,.2f}'}")
    print("--------------------------------------------------")
```

```python
#3.4
import random
import time
from datetime import datetime

# --- Base Class ---
class TrafficLight:
    def __init__(self, location):
        self.location = location
        self.__current_state = "Red"  # private attribute (encapsulation)

    def turn_green(self):
        self.__current_state = "Green"

    def turn_red(self):
        self.__current_state = "Red"

    def get_state(self):
        return self.__current_state


#Simulated Sensor API
class SensorAPI:
    @staticmethod
    def get_car_density(location: str) -> int:
        random.seed(hash(location) ^ random.randint(0, 10**6))
        return random.randint(0, 120)  # cars per minute


#Subclass: SmartTrafficLight
class SmartTrafficLight(TrafficLight):
    def __init__(self, location, cycle_total=90, min_green=30, max_green=60):
        super().__init__(location)
        self.cycle_total = cycle_total
        self.min_green = min_green
        self.max_green = max_green
        self.green_duration = min_green
        self.red_duration = cycle_total - self.green_duration
        self.last_density = None

    def poll_sensor(self):
        """Fetch simulated car density."""
        self.last_density = SensorAPI.get_car_density(self.location)
        return self.last_density

    def adjust_timings(self):
        """Adjust green/red duration based on car density."""
        if self.last_density is None:
            self.poll_sensor()

        d = max(0, min(120, self.last_density))
        span = self.max_green - self.min_green
        self.green_duration = int(self.min_green + span * (d / 120.0))
        self.red_duration = self.cycle_total - self.green_duration
        return self.green_duration, self.red_duration

    def simulate_cycle(self):
        """Simulate one traffic light cycle and print log."""
        self.poll_sensor()
        green, red = self.adjust_timings()

        # Turn green
        self.turn_green()
        timestamp = datetime.now().strftime("[%H:%M]")
        print(f"{timestamp} {self.location}: Green for {green}s, Red for {red}s.")

        # For simulation, just pause briefly instead of real-time waiting
        time.sleep(0.5)  # simulate time passing quickly

        # Turn red
        self.turn_red()


# Simulation of multiple cycles 
if __name__ == "__main__":
    light = SmartTrafficLight("Jinja Road")

    print("=== TRAFFIC LIGHT SIMULATION START ===")
    for i in range(5): #5 cycles
        light.simulate_cycle()
    print("=== SIMULATION END ===")
```

```python
#4.4
import json

#Base Class: Student
class Student:
    def __init__(self, student_id, name, marks):
        self.student_id = student_id
        self.name = name
        self.marks = marks

    def compute_total(self):
        return sum(self.marks.values())

    def compute_average(self):
        if not self.marks:
            return 0
        return self.compute_total() / len(self.marks)

    def get_grade(self):
        avg = self.compute_average()
        if avg >= 80:
            return "A"
        elif avg >= 70:
            return "B"
        elif avg >= 60:
            return "C"
        elif avg >= 50:
            return "D"
        else:
            return "F"

    def to_dict(self):
        """Convert student data to dictionary for JSON serialization."""
        return {
            "student_id": self.student_id,
            "name": self.name,
            "marks": self.marks,
            "total": self.compute_total(),
            "average": self.compute_average(),
            "grade": self.get_grade()
        }

    @classmethod
    def from_dict(cls, data):
        """Recreate Student object from dictionary."""
        return cls(data["student_id"], data["name"], data["marks"])

    def display_info(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        for subject, mark in self.marks.items():
            print(f"  {subject}: {mark}")
        print(f"Total Marks: {self.compute_total()}")
        print(f"Average Marks: {self.compute_average():.2f}")
        print(f"Grade: {self.get_grade()}")


#Subclass: PostgraduateStudent
class PostgraduateStudent(Student):
    def __init__(self, student_id, name, marks, research_topic):
        super().__init__(student_id, name, marks)
        self.research_topic = research_topic

    def evaluate_thesis(self, score):
        return "Pass" if score >= 70 else "Revise"

    def to_dict(self):
        """Extend parent dictionary with research topic."""
        base_data = super().to_dict()
        base_data["research_topic"] = self.research_topic
        base_data["type"] = "PostgraduateStudent"
        return base_data

    @classmethod
    def from_dict(cls, data):
        return cls(data["student_id"], data["name"], data["marks"], data["research_topic"])


# Save and Load Functions
def save_students_to_json(students, filename="students.json"):
    """Save a list of Student/PostgraduateStudent objects to JSON file."""
    with open(filename, "w", encoding="utf-8") as f:
        json.dump([s.to_dict() for s in students], f, indent=4)
    print(f"Data saved to {filename}.")


def load_students_from_json(filename="students.json"):
    """Load students from JSON and recreate objects."""
    with open(filename, "r", encoding="utf-8") as f:
        data_list = json.load(f)

    students = []
    for data in data_list:
        if data.get("type") == "PostgraduateStudent":
            students.append(PostgraduateStudent.from_dict(data))
        else:
            students.append(Student.from_dict(data))
    print(f"Data loaded from {filename}.")
    return students
```

```python
#4.5
class Student:
    def __init__(self, student_id, name, marks):
        self.student_id = student_id
        self.name = name
        self.marks = marks

    def compute_total(self):
        return sum(self.marks.values())

    def compute_average(self):
        return 0 if not self.marks else self.compute_total() / len(self.marks)

    def get_grade(self):
        avg = self.compute_average()
        return "A" if avg >= 80 else "B" if avg >= 70 else "C" if avg >= 60 else "D" if avg >= 50 else "F"

    # --- Polymorphic interface ---
    def evaluation_report(self, thesis_score=None):
        return {
            "student_id": self.student_id,
            "name": self.name,
            "coursework_total": self.compute_total(),
            "coursework_average": round(self.compute_average(), 2),
            "coursework_grade": self.get_grade(),
            "thesis_result": None  # base students have no thesis
        }


class PostgraduateStudent(Student):
    def __init__(self, student_id, name, marks, research_topic):
        super().__init__(student_id, name, marks)
        self.research_topic = research_topic

    def evaluate_thesis(self, score):
        return "Pass" if score >= 70 else "Revise"

    # Polymorphic override: adds thesis evaluation
    def evaluation_report(self, thesis_score=None):
        base = super().evaluation_report(thesis_score)
        base["research_topic"] = self.research_topic
        base["thesis_result"] = None if thesis_score is None else self.evaluate_thesis(thesis_score)
        return base


# Demo of polymorphism
students = [
    Student("S001", "Amina K", {"Math": 78, "Stats": 84, "Python": 90}),
    PostgraduateStudent("PG002", "Ken O", {"Methods": 82, "Data": 79, "ML": 88}, "PDM Allocation Models")
]

for s in students:
    report = s.evaluation_report(thesis_score=75)
    print(report)
```

```python
#6.2
import json
import requests
import pandas as pd
from datetime import datetime


class ExchangeRatePipeline:
    def __init__(self, json_path: str = None, api_url: str = None):
        """
        Initialize pipeline with either a local file or API endpoint.
        """
        self.json_path = json_path
        self.api_url = api_url
        self.raw_data = None
        self.df = None

    def fetch(self):
        """Fetch data from API or load from local JSON file."""
        if self.api_url:
            try:
                response = requests.get(self.api_url, timeout=10)
                response.raise_for_status()
                self.raw_data = response.json()
                print("✅ Data fetched successfully from API.")
            except requests.RequestException as e:
                print(f"⚠️ Network or API error: {e}")
                self.raw_data = None
        elif self.json_path:
            try:
                with open(self.json_path, "r", encoding="utf-8") as f:
                    self.raw_data = json.load(f)
                print("✅ Data loaded successfully from local JSON.")
            except FileNotFoundError:
                print(f"❌ File not found: {self.json_path}")
            except json.JSONDecodeError as e:
                print(f"⚠️ JSON decoding error: {e}")
        else:
            print("⚠️ No data source provided (API or JSON file).")

        return self.raw_data

    def clean_transform(self):
        """
        Clean and transform the raw exchange rate data into a pandas DataFrame.
        Expected JSON structure:
        {
          "date": "2025-10-31",
          "rates": { "USD": 3479.78, "GBP": 4629.88, "EUR": 4045.11 }
        }
        """
        if not self.raw_data:
            print("⚠️ No data to clean — run fetch() first.")
            return None

        try:
            # Convert single-day JSON to DataFrame 
            if isinstance(self.raw_data, dict) and "rates" in self.raw_data:
                df = pd.DataFrame([{
                    "date": pd.to_datetime(self.raw_data.get("date", datetime.today())),
                    "USD": float(self.raw_data["rates"].get("USD", None)),
                    "GBP": float(self.raw_data["rates"].get("GBP", None)),
                    "EUR": float(self.raw_data["rates"].get("EUR", None))
                }])
            elif isinstance(self.raw_data, list):
                df = pd.DataFrame([
                    {
                        "date": pd.to_datetime(item.get("date")),
                        "USD": float(item["rates"].get("USD", None)),
                        "GBP": float(item["rates"].get("GBP", None)),
                        "EUR": float(item["rates"].get("EUR", None))
                    }
                    for item in self.raw_data if "rates" in item
                ])
            else:
                print("⚠️ Unexpected JSON format.")
                return None

            # --- Cleaning and transformations ---
            df = df.dropna(subset=["USD", "GBP", "EUR"])   # remove rows with missing rates
            df = df.sort_values(by="date").reset_index(drop=True)
            df["Month"] = df["date"].dt.strftime("%Y-%m")

            self.df = df
            print("✅ Data cleaned and transformed successfully.")
            return self.df

        except Exception as e:
            print(f"⚠️ Error during cleaning: {e}")
            return None

    def summary(self):
        """Return summary statistics for each currency."""
        if self.df is None:
            print("⚠️ No cleaned DataFrame found — run clean_transform() first.")
            return None

        summary_df = self.df[["USD", "GBP", "EUR"]].describe()
        print("📊 Exchange Rate Summary (UGX per currency):")
        print(summary_df)
        return summary_df

    def save_to_csv(self, filename="clean_exchange_rates.csv"):
        """Save cleaned data to CSV file."""
        if self.df is None:
            print("⚠️ No data to save — run clean_transform() first.")
            return
        self.df.to_csv(filename, index=False)
        print(f"✅ Cleaned exchange rates saved to {filename}")
```

```python
#6.6
import json
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime


class ExchangeRatePipeline:
    def __init__(self, json_path: str = None, api_url: str = None):
        self.json_path = json_path
        self.api_url = api_url
        self.raw_data = None
        self.df = None

    #FETCH
    def fetch(self):
        """Fetch data from API or local JSON file with HTTP/IO error handling."""
        if self.api_url:
            try:
                res = requests.get(self.api_url, timeout=15)
                res.raise_for_status()
                self.raw_data = res.json()
                if not self.raw_data:
                    raise ValueError("Empty JSON payload from API.")
                print("✅ Data fetched from API.")
            except requests.exceptions.HTTPError as e:
                print(f"❌ HTTP error: {e} (status={getattr(e.response, 'status_code', 'n/a')})")
                self.raw_data = None
            except requests.exceptions.Timeout:
                print("❌ Request timed out.")
                self.raw_data = None
            except requests.exceptions.RequestException as e:
                print(f"❌ Network error: {e}")
                self.raw_data = None
            except ValueError as e:
                print(f"❌ JSON error: {e}")
                self.raw_data = None
        elif self.json_path:
            try:
                with open(self.json_path, "r", encoding="utf-8") as f:
                    self.raw_data = json.load(f)
                if not self.raw_data:
                    raise ValueError("Empty JSON file.")
                print("✅ Data loaded from local JSON.")
            except FileNotFoundError:
                print(f"❌ File not found: {self.json_path}")
                self.raw_data = None
            except json.JSONDecodeError as e:
                print(f"❌ JSON decode error: {e}")
                self.raw_data = None
            except ValueError as e:
                print(f"❌ Data error: {e}")
                self.raw_data = None
        else:
            print("❌ No data source provided (api_url or json_path).")
            self.raw_data = None

        return self.raw_data

    #CLEAN & TRANSFORM
    def clean_transform(self):
        """
        Clean and transform exchange rate data into a pandas DataFrame.
        Expected shapes:
        - dict: {"date": "...", "rates": {"USD":..., "GBP":..., "EUR":...}}
        - list[dict]: same structure repeated over time
        """
        if not self.raw_data:
            print("❌ No data available. Run fetch() first.")
            return None

        try:
            records = []

            def _extract(record):
                if "rates" not in record:
                    raise KeyError("Missing 'rates' in record.")
                date_val = record.get("date")
                if not date_val:
                    raise KeyError("Missing 'date' in record.")
                usd = record["rates"].get("USD")
                gbp = record["rates"].get("GBP")
                eur = record["rates"].get("EUR")
                if usd is None or gbp is None or eur is None:
                    raise KeyError("Missing one of USD/GBP/EUR in 'rates'.")
                return {
                    "date": pd.to_datetime(date_val),
                    "USD": float(usd),
                    "GBP": float(gbp),
                    "EUR": float(eur),
                }

            if isinstance(self.raw_data, dict):
                records.append(_extract(self.raw_data))
            elif isinstance(self.raw_data, list):
                for item in self.raw_data:
                    try:
                        records.append(_extract(item))
                    except (KeyError, ValueError) as e:
                        # Skip bad rows but continue
                        print(f"⚠️ Skipping bad record: {e}")
            else:
                raise TypeError("Unexpected JSON structure (not dict or list).")

            if not records:
                raise ValueError("No valid records after cleaning.")

            df = pd.DataFrame(records).dropna(subset=["USD", "GBP", "EUR"])
            if df.empty:
                raise ValueError("All rows invalid or missing required rates.")

            df = df.sort_values("date").reset_index(drop=True)
            df["Month"] = df["date"].dt.to_period("M").astype(str)
            self.df = df
            print("✅ Data cleaned and transformed successfully.")
            return df

        except (KeyError, ValueError, TypeError) as e:
            print(f"❌ Cleaning error: {e}")
            self.df = None
            return None

    #MONTHLY AVERAGES
    def monthly_average(self):
        """Compute monthly average exchange rates using NumPy with safeguards."""
        if self.df is None or self.df.empty:
            print("❌ No DataFrame available — run clean_transform() first.")
            return None

        try:
            months = self.df["Month"].unique()
            averages = []
            for m in months:
                subset = self.df[self.df["Month"] == m]
                if subset.empty:
                    continue
                averages.append({
                    "Month": m,
                    "USD_avg": float(np.mean(subset["USD"].to_numpy())),
                    "GBP_avg": float(np.mean(subset["GBP"].to_numpy())),
                    "EUR_avg": float(np.mean(subset["EUR"].to_numpy()))
                })
            if not averages:
                raise ValueError("No monthly aggregates could be computed.")
            avg_df = pd.DataFrame(averages)
            print("📊 Monthly average exchange rates computed.")
            return avg_df
        except Exception as e:
            print(f"❌ Monthly average error: {e}")
            return None

    #PLOT TRENDS
    def plot_trends(self, use_monthly_averages: bool = False):
        """Plot exchange rate trends; guard against missing data."""
        if self.df is None or self.df.empty:
            print("❌ No data to plot. Run clean_transform() first.")
            return
        try:
            if use_monthly_averages:
                df = self.monthly_average()
                if df is None or df.empty:
                    print("❌ No monthly averages to plot.")
                    return
                x = df["Month"]; usd, gbp, eur = df["USD_avg"], df["GBP_avg"], df["EUR_avg"]
                title = "Monthly Average Exchange Rates (UGX per unit)"
                x_label = "Month"
            else:
                df = self.df
                x = df["date"]; usd, gbp, eur = df["USD"], df["GBP"], df["EUR"]
                title = "Daily Exchange Rates (UGX per unit)"
                x_label = "Date"

            plt.figure(figsize=(10, 6))
            plt.plot(x, usd, label="USD/UGX", marker="o")
            plt.plot(x, gbp, label="GBP/UGX", marker="s")
            plt.plot(x, eur, label="EUR/UGX", marker="^")
            plt.title(title); plt.xlabel(x_label); plt.ylabel("UGX")
            plt.grid(alpha=0.3); plt.legend(); plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"❌ Plotting error: {e}")

    #EXPORT DATA
    def export_data(self, base_filename="exchange_rates"):
        """Export daily and monthly data to CSV/JSON with error handling."""
        if self.df is None or self.df.empty:
            print("❌ No processed data available — run clean_transform() first.")
            return

        try:
            # Daily
            csv_file = f"{base_filename}_daily.csv"
            json_file = f"{base_filename}_daily.json"
            self.df.to_csv(csv_file, index=False)
            self.df.to_json(json_file, orient="records", indent=4, date_format="iso")

            # Monthly
            avg_df = self.monthly_average()
            if avg_df is None or avg_df.empty:
                print("⚠️ Monthly averages not exported (no data).")
            else:
                avg_csv = f"{base_filename}_monthly_avg.csv"
                avg_json = f"{base_filename}_monthly_avg.json"
                avg_df.to_csv(avg_csv, index=False)
                avg_df.to_json(avg_json, orient="records", indent=4)

            print("✅ Export completed.")
        except Exception as e:
            print(f"❌ Export error: {e}")
```
